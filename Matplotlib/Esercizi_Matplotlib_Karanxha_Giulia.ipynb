```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

#Es 1

```python
x = np.linspace(-5,5,100)
y = 2*x+1
plt.figure(figsize=(15,8))
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2843', )
plt.ylabel('y', color='#1C2843')
plt.legend(loc='upper left')
#plt.grid()
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-np.pi,np.pi,100)
# the function, which is y = sin(x) here #funzione seno perché per sua natura è compresa tra 0 ed 1/-1
y = np.sin(x) #funzione periodica ed è seno perché il seno di 0 è 0; è come una parabola che ogni volta si rovescia ad ogni pi greco
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'b')
# show the plot
plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(-np.pi,np.pi,100)
# the functions, which are y = ? and z = ? here
y = np.sin(x)
z = np.cos(x)
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the functions
plt.plot(x,y, 'c', label='y=sin(x)')
plt.plot(x,z, 'm', label='z=cos(x)')
plt.legend(loc='upper left')
# show the plot
plt.show()
```

```python

# 100 linearly spaced numbers
x = np.linspace(-2,2,100)
# the function, which is y = e^x here
y = np.exp(x) #esponenziale
# setting the axes at the centre
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
# show the plot
plt.show();
```

```python

# 100 linearly spaced numbers
x = np.linspace(-2,2,100)
# the function, which is y = e^x here
y = np.exp(x**3) #esponenziale, più è alto l'argomento della funzione esponenziale, più aumenta velocemente
# setting the axes at the centre
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
# show the plot
plt.show();
```
