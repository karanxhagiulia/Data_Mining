#Funzioni ricorsive

Si autorichiama nel corpo della funzione stessa

Esempio:


````python
#calcola il fattoriale di un numero
#Es, !5 = 5x4x3x2x1 (la moltiplicazione di tutti i numeri interi dal numero ad uno), cresce più velocemente dell'esponenziale, il fattoriale di 0 è 1, !0 = 1, !1 = 1

def fatt(n):
  if n < 2:
    return 1
  else:
    return n * fatt(n-1) #il risultato di n>=2 è n* il fattoriale del valore di n precedente ( ovvero n-1 a meno che non sia <2)

fatt(3) #oppure
print(fatt(5))
````

Senza ricorsione:

````python
#senza ricorsione

def fatt2(n):
  f = n
  for i in range(1,n): #devo arrivare a n compreso, gli moltiplico 1 2 3 4 fino a n
    f *= i # f(ovvero n) è uguale a f (ovvero n) moltiplicato per i nella lunghezza di n 
  return f

fatt2(5)

# oppure scrivendo di più:

#f= 1
#for i in range(2, n+1)
````
